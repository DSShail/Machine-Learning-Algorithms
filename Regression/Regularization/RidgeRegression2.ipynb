```python
#for creating custom dataset
from sklearn.datasets import make_regression

import matplotlib.pyplot as plt
import numpy as np
```

### make_regression

- generates a random regression problem

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
plt.scatter(X,y)
```

### Let's apply linear regression over the dataset

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X,y)
print("Coefficient: ",lr.coef_)
print("Intercept: ",lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=10)
rr.fit(X,y)
print("Coefficient: ",rr.coef_)
print("Intercept: ",rr.intercept_)
```

##### As we can see that the  value of the slope has decreased from 27 to 24 after applying ridge regression

```python
from sklearn.linear_model import Ridge
rr1=Ridge(alpha=100)
rr1.fit(X,y)
print("Coefficient: ",rr1.coef_)
print("Intercept: ",rr1.intercept_)
```

```python

plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.plot(X,rr1.predict(X),color='orange',label='alpha=100')
plt.legend()
```

#### See the impact of alpha value using ridge regression in above figure
